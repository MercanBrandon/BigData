3) En créant un nuage de points, regardez comment se comporte la colonne LotArea par rapport au SalesPrices

```python
plt.pyplot.scatter(data['LotArea'], data['SalePrice'])
```

4) Affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotArea inférieur à 20 000 pieds carrés et un prix inférieur à 500 000$

```python
# x = data[data['LotArea'] < 20000]
# y = data[data['SalePrice']< 500000]



x = data['LotArea']
y = data['SalePrice']
axes = plt.pyplot.gca()
axes.set_xlim(0,20000)
axes.set_ylim(0,500000)
axes.set_xlabel('')
axes.set_ylabel('Lot Area')
axes.scatter(x,y)

```

```python
En créant un nuage de points, regardez la relation entre le LotFrontage et le LotArea
```

```python
x = data['LotFrontage']
y = data['LotArea']

axes = plt.pyplot.gca()
axes.set_xlabel('Lot Area')
axes.set_ylabel('Sales Prices')
axes.scatter(x,y)
```

6) De la même manière, affinez votre visualisation en ne gardant uniquement les maisons

```python
axes = plt.pyplot.gca()
axes.set_xlabel('Lot Frontage')
axes.set_ylabel('Lot Area')
axes.set_xlim(0,200)
axes.set_ylim(0,100000)
axes.scatter(x,y)
```
